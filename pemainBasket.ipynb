## - Hunting Pemain Basket -

<hr>

### Kriteria yang dipakai

- Usia (Age) <= 25 tahun,
- Tinggi badan (Height) >= 180 cm,
- Berat badan (Weight) <= 90 kg,
- Rata-rata point (Average points scored) >= 6, dan
- Rata-rata rebound (Average rebounds grabbed) >= 3.

<hr>

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
nbaDf = pd.read_csv('all_seasons.csv')
nbaDf.head()
```

```python
# Cek NA datasets
nbaDf.isnull().sum()
```

```python
# Sortir parameter yang dipakai
newDf = nbaDf[['age', 'player_height', 'player_weight', 'pts', 'reb']]
newDf
```

### Membuat DataFrame pemain Indonesia

```python
IndoDf = pd.DataFrame({
    'Name': ['Andakara Prastawa Dyaksa', 'Reggie Mononimbar', 'Hardianus Lakudu', 'Kevin Yonas Sitorus', 'Arki Dikania Wisnu', 'Laurentius Steven Oei', 'Mei Joni', 'Vincent Rivaldi Kosasih', 'Hardian Wicaksono', 'Brandon Jawato'],
    'Club': ['Pelita Jaya Bakrie', 'Pelita Jaya Bakrie', 'Satria Muda Pertamina Jakarta', 'Satria Muda Pertamina Jakarta', 'Satria Muda Pertamina Jakarta', 'Satria Muda Pertamina Jakarta', 
             'Stapac', 'Stapac', 'Pacific Caesar Surabaya', 'Louvre Surabaya'],
    'Nationality': ['Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia', 'Indonesia'],
    'Age': [24, 21, 23, 26, 20, 21, 25, 23, 21, 24],
    'Height': [190, 185, 178, 185, 183, 191, 188, 179, 177, 182],
    'Weight': [90, 86, 83, 75, 80, 85, 90, 87, 80, 85],
    'Avg_points': [7, 6, 10, 11, 5, 4, 7, 1, 9, 6],
    'Avg_rebounds': [6, 3, 3, 4, 2, 10, 5, 2, 8, 5]
})

IndoDf
```

### - Membuat target sesuai kriteria -

```python
# Filter pemain NBA sesuai kriteria
newDf['Target'] = newDf.apply(lambda x: 1 if (x['age'] <= 25) & (x['player_height'] >= 180) & (x['player_weight'] <= 90) & (x['pts'] >= 6) & (x['reb'] >= 3) else 0, axis=1)
newDf
```

### - Train test split dan Standardize -

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x = newDf[['age','player_height','player_weight', 'pts', 'reb']]
y = newDf['Target']

scaler = StandardScaler()
scaler.fit(x)
z = scaler.transform(x)
print(z)

xtr, xts, ytr, yts = train_test_split(x,y, test_size=.18, random_state=100)
print(xtr.shape)
```

<hr>

## Model Machine Learning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
from sklearn import tree
```

### - LogisticRegression -

```python
# param yang akan dituned LogisticRegression
penalty = ['l1', 'l2', 'elasticnet', 'none']
solver  = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
max_iter = [1, 10, 100, 1000, 10000]

param1 = {
    'penalty': penalty, 'solver': solver, 'max_iter': max_iter
}
```

```python
model   = LogisticRegression()
modelrs = RandomizedSearchCV(
    estimator = model, param_distributions = param1
)
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
modelrs.fit(xtr, ytr)
modelrs.best_params_
```

```python
model1 = LogisticRegression()
model1.fit(xtr, ytr)

model1Baru = LogisticRegression(solver='liblinear', penalty='l1', max_iter=1000)
model1Baru.fit(xtr, ytr)

print(f'LogisticRegression asli  : {model1.score(xts, yts)}')
print(f'LogisticRegression tuning: {model1Baru.score(xts, yts)}')
```

### - Random Forest -

```python
# param yang akan dituned Random Forest

n_estimators = [1,10,100,1000]
criterion    = ['gini','entropy']
bootstrap    = [True,False]
         

param2 = {'criterion' : criterion,
          'bootstrap' : bootstrap,
          'n_estimators': n_estimators
         }
```

```python
model2   = RandomForestClassifier()
modelrs2 = RandomizedSearchCV(
    estimator = model2, param_distributions = param2
)
```

```python
modelrs2.fit(xtr, ytr)
modelrs2.best_params_
```

```python
model2 = RandomForestClassifier()
model2.fit(xtr, ytr)

model2Baru = RandomForestClassifier(
    n_estimators=10, criterion='gini', bootstrap=False
)
model2Baru.fit(xtr, ytr)

print(f'RandomForest asli  : {model2.score(xts, yts)}')
print(f'RandomForest tuning: {model2Baru.score(xts, yts)}')
```

### - Decision Tree -

```python
param3 = {'criterion' : ['gini', 'entropy'], 'splitter' : ['best','random'], 'random_state' : [2,4,6,8,10]}
```

```python
model3   = tree.DecisionTreeClassifier()
modelrs3 = RandomizedSearchCV(
    estimator = model3, param_distributions = param3
)
```

```python
modelrs3.fit(xtr, ytr)
modelrs3.best_params_
```

```python
model3 = tree.DecisionTreeClassifier()
model3.fit(xtr, ytr)

model3Baru = tree.DecisionTreeClassifier(splitter= 'best', criterion= 'gini', random_state=4)
model3Baru.fit(xtr, ytr)

print(f'DecisionTree asli  : {model3.score(xts, yts)}')
print(f'DecisionTree tuning: {model3Baru.score(xts, yts)}')
```

## Evaluation Matrics
Balanced Accuracy, Precision, Recall, F1 score & ROC-AUC score

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, roc_auc_score
```

- Logistic Regression

```python
y_pred_log = model1.predict(xts)
print('Balanced-Acc : ', round(balanced_accuracy_score(yts, y_pred_log)*100,2), '%')
print('Precision(+) : ', round(precision_score(yts, y_pred_log, pos_label=1)*100,2), '%')
print('Precision(-) : ', round(precision_score(yts, y_pred_log, pos_label=0)*100,2), '%')
print('Recall(+)    : ', round(recall_score(yts, y_pred_log, pos_label=1)*100,2), '%')
print('Recall(-)    : ', round(recall_score(yts, y_pred_log, pos_label=0)*100,2), '%')
print('F1 score(+)  : ', round(f1_score(yts, y_pred_log, pos_label=1)*100,2), '%')
print('F1 score(-)  : ', round(f1_score(yts, y_pred_log, pos_label=0)*100,2), '%')
print('Nilai AUC    : ', round(roc_auc_score(yts, y_pred_log)*100,2), '%')
```
